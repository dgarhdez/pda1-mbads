### Exercise 11

In the previous text, how many words are duplicates?

````python
type(text.replace("(", ""))
````

````python
# removing special character that are not numbers or letters
text_replace_spec_chars = (
    text.replace("(", "").replace(")", "").replace(",", "").replace(".", "")
)

# convert to uppercase
text_replace_spec_chars_upper = text_replace_spec_chars.upper()

# elements in the list after splitting
n_words_list = len(text_replace_spec_chars_upper.split())

# convert list form string into set: to remove duplicates
unique_words = set(text_replace_spec_chars_upper.split())

# count elements in set
n_unique_words = len(unique_words)

# number of duplicated words
duplicates = n_words_list - n_unique_words

n_words_list, n_unique_words, duplicates
````
